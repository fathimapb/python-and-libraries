#Numpy filter array

```python
import numpy as np
arr=np.array([41,42,43,44])
x=[True,False,True,False]
newarr=arr[x]
print(newarr)
```

##Creating the filter array

```python
#Create a filter array that will return only values higher than 42:
arr=np.array([41,42,43,44])
filter_arr=[]
for element in arr:
  if element>42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
#create a filter array that will return only even elements from the original array:
arr=np.array([1,2,3,4,5,6,7])
filter_arr=[]
for element in arr:
  if element % 2 ==0:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

##Creatin filter directly from array

```python
# create a filter array that will return only values higher than 42:
arr=np.array([41,42,43,44])
filter_arr=arr>42
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
# create a filter array that will return only even elements fron the original array:
arr=np.array([1,2,3,4,5,6,7])
filter_arr=arr % 2==0
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

#InterQuartile Range(IQR)

```python
import numpy as np
import scipy.stats as stats

data=[12,15,14,10,18,20,22,25,14,17]

range_value=max(data)-min(data)
print(f"Range:{range_value}")
```

```python
variance_value=np.var(data,ddof=0)
sample_variance_value=np.var(data,ddof=1)
print(f"population variance:{variance_value}")
print(f"sample variance:{sample_variance_value}")
```

```python
std_dev_value=np.std(data,ddof=0)
sample_std_dev_value=np.std(data,ddof=1)
print(f"population standard deviation:{std_dev_value}")
print(f"sample standard deviation:{sample_std_dev_value}")
```

```python
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
iqr_value=Q3-Q1
print(f"Interquartile Range(IQR):{iqr_value}")
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
newarr=np.add(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([1,2,3])
newarr=np.sum([arr1,arr2])
print(newarr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([1,2,3])
newarr=np.sum([arr1,arr2],axis=1)
print(newarr)
```

```python
arr=np.array([1,2,3])
newarr=np.cumsum(arr)
print(newarr)
```

```python
arr=np.array([1,2,3,4])
x=np.prod(arr)
print(x)
```

```python
arr1=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
x=np.prod([arr1,arr2])
print(x)
```

```python
arr1=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
x=np.prod([arr1,arr2],axis=1)
print(x)
```

```python
arr=np.array([5,6,7,8])
newarr=np.cumprod(arr)
print(newarr)
```

```python
arr=np.array([10,15,25,5])
newarr=np.diff(arr)
print(newarr)
```

```python
arr=np.array([10,15,25,5])
newarr=np.diff(arr,n=2)
print(newarr)
```

#NumPy LCM Lowest Commom Multiple

```python
num1=15
num2=87
x=np.lcm(num1,num2)
print(x)
```

```python
arr=np.array([1,2,3,4])
x=np.lcm.reduce(arr)
print(x)
```

```python
arr=np.arange(1,10)
x=np.lcm.reduce(arr)
print(x)
```

#NumPy GCD Greatest Commom Denominator

```python
num1=6
num2=9
x=np.gcd(num1,num2)
print(x)
```

#NumPy Set Operation

```python
arr=np.array([1,1,1,2,3,4,5,5,6,7])
x=np.unique(arr)
print(x)
```

```python
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
x=np.intersect1d(arr1,arr2,assume_unique=True)
print(x)
```

```python
set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])
x=np.setdiff1d(set1,set2,assume_unique=True)
print(x)
```

```python
set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])
x=np.setxor1d(set1,set2,assume_unique=True)
print(x)
```
