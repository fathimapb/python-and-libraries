##Create DataFrame Using Dictionary

```python
import pandas as pd
emp={'Id':[101,102,103,104,105],
     'Name':['Deepak','Shyam','Arun','Manu','Jeena'],
     'Age':[25,22,20,23,21],
     'Salary':[25000,30000,15000,45000,10000]}

df=pd.DataFrame(emp)
df

```
