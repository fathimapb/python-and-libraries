##Create a tile for a plot

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title("Sports watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

##Set font properties for title and labels

```python
# we can use the fontdict parameter in xlabel(),ylabel() and title() to set font properties for the title and label
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}

plt.title("Sports watch Data",fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calorie Burnage",fontdict=font2)

plt.plot(x,y)
plt.show()
```

##Position the title

##Title
you can add a title to each plot with the title() function

```python
# plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")
# plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")
plt.show()
```

##Super Title
you can add a title to the entire figure with the suptitle() function

```python
# plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")
# plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")
plt.suptitle("MY SHOP")
plt.show()
```

##Size

##Combine Color Size and Alpha

```python
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()

```

#Matplotlib Pie Charts

```python
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

##Labels

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels=mylabels)
plt.show()
```

##Start Angle

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

##Explode

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0,0.2,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

##Shadow

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0,0.2,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

##Colors

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
mycolors=["black","hotpink","b","#4CAF50"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

##Legend

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

##Legend with header

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]

plt.pie(y,labels=mylabels)
plt.legend(title="Fouer Fruits: " )
plt.show()
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=10,
    values=[48,46,6],
    figsize=(5,3)
)
plt.show()
```

```python
# Values in dict&Auto-sizing
data={'Democratic':48,'Republician':46,'Libertarian':3}

fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    legend={'loc':'upper left','bbox_to_anchor':(0,-0.4)}
)
plt.show()
```

```python
data={'Democratic':48,'Republician':46,'Libertarian':3}

fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=['#232066','#883D3D','#DCB732'],
    title={'label':'Vote Percentage in 2016 US Presidential Election','loc':'left'},
    labels=[f"{k}({v}%6)"for k,v in data.items()],
    legend={'loc':'upper left','bbox_to_anchor':(0,-0.4),'ncol':len(data),'framealpha':0},
    starting_location='NW',
    block_arranging_style='snake'
)
fig.set_facecolor('#EEEEEE')
plt.show()
```

```python
data={'Democratic':48,'Republician':46,'Libertarian':3}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=['#232066','#3503fc','#DCB732'],
    legend={'loc':'upper left','bbox_to_anchor':(1,1)},
    icons='child',
    font_size=12,
    icon_legend=True
)
plt.show()
```
