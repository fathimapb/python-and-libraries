```python
from abc import ABC , abstractmethod

class Shape(ABC):

  @abstractmethod
  def calculate_area(self):
    pass

class circle(Shape):

  def __init__(self,radius):
    self.radius=radius

  def calculate_area(self):
    print(f"circle area is: { 3.14 * (self.radius) **2}")

class Rectangle(Shape):

  def __init__(self,width,height):
    self.width=width
    self.height=height

  def calculate_area(self):
    print(f"Rectangle area is: {self.width * self.height}")


c=circle(10)
r=Rectangle(5,10)

c.calculate_area()
print()
r.calculate_area()



```

Student Management System

Extend your Student class to include multiple students in a Classroom.

Methods to:

Add a student

Display all students

Find the student with the highest average.



```python
class Student:

  def __init__(self):
    self.name=[]
    self.mark=[]

  def add_student(self,name,mark):
    self.name.append(name)
    self.mark.append(mark)


  def display_student(self):
    print(f"Total students in the class: {self.name}")

  def get_average(self):
    if len(self.mark)==0:
      print()
    total=0
    for mark in self.mark:
      total+=mark
    print(total/len(self.mark))



s=Student()
s.add_student("fathima",34)
s.add_student("Nasila",54)
s.add_student("Ameena",33)
s.add_student("Fizza",12)

s.display_student()
s.get_average()

```

ATM Simulation

Create a class ATM that allows users to log in with a PIN.

Supports deposit, withdrawal, balance inquiry.

Add error handling for invalid PIN, insufficient funds, etc.

```python
class ATM:

  def __init__(self,pin):
    self.pin=pin
    self.balance=6000


  def deposit(self,pin,amount):
    if pin==self.pin:
      print("Succesfully logged in")
      if amount>0:
        self.balance+=amount
        print(f"deposit {amount},balance:{self.balance}")
      else:
        print("invalid amount")
    else:
      print("Incorrect pin!!")

  def withdraw(self,pin,amount):
    if pin==self.pin:
      print("Succesfully logged in")
      if amount < self.balance:
        self.balance-=amount
        print(f"withdraw {amount},balance:{self.balance}")
    else:
      print("Incorrect pin!!")


  def balance_inquiry(self):
    print(f"current balance:{self.balance}")


a=ATM(1234)
a.deposit(1234,5600)
a.withdraw(1234,3000)
a.balance_inquiry()

```

```python
import threading
import time

def download_file(file):
  print(f"Downloading{file}....")
  time.sleep(2)
  print(f"{file} downloaded")

files=["file1.txt","file2.txt","file3.txt"]
thread=[threading.Thread(target=download_file,args=(f,))for f in files]

for t in thread:
  t.start()

for t in thread:
  t.join()
```

```python
import threading
import time

def chef_work():
  for i in range(5):
    print(f"chef: Preparing dish {i + 1}....")
    time.sleep(2)
    if i==2:
      print("chef: Calling assistant to wash utensils.")
      time.sleep(1)
      assistant_thread=threading.Thread(target=assistant_work)
      assistant_thread.start()
      assistant_thread.join()
      print("chef: Assistant done.Resuming dish preparation.")
  print("chef: All dishes are prepared.")


def assistant_work():
  for j in range(3):
    print(f"Assistant: washing utensil {j + 1}....")
    time.sleep(2)
  print("Assistant: All utensils are clean.")


print("kitchen: starting work.")
chef_thread=threading.Thread(target=chef_work)
chef_thread.start()
chef_thread.join()
print("kitchen: All work completed.")
```

##Question
Write a Python program using threading to simulate music playback and lyrics downloading:

Music playback runs in one thread for 5 seconds, printing a message each second.
During the 3rd second, a separate thread starts to simulate downloading lyrics, which takes 3 seconds.
Ensure the music playback waits for the lyrics to finish downloading before continuing.



**Output**

Music Player: Starting music playback.
Music Player: Playing music... 1 seconds
Music Player: Playing music... 2 seconds
Music Player: Playing music... 3 seconds
Music Player: Downloading lyrics in the background.
  Downloading lyrics... 1 seconds
  Downloading lyrics... 2 seconds
  Downloading lyrics... 3 seconds
  Lyrics download completed.
Music Player: Lyrics downloaded.
Music Player: Playing music... 4 seconds
Music Player: Playing music... 5 seconds
Music Player: Playback ended

```python
import threading
import time

def music_playback():
  for i in range(5):
    print(f"Music player: Playing music {i + 1} seconds")
    time.sleep(2)
    if i==2:
      print("Music player: Downloading lyrics in the background")
      time.sleep(1)
      download_thread=threading.Thread(target=downloading_lyrics)
      download_thread.start()
      download_thread.join()
      print("lyric download completed")


def downloading_lyrics():
  for j in range(3):
    print(f"Downloading lyrics: Downloading lyrics {j + 1} seconds")
    time.sleep(2)
  print("Music player: lyric downloaded")

print("Music player: Starting music playback")
music_thread=threading.Thread(target=music_playback)
music_thread.start()
music_thread.join()
print("Music player: Playback ended")


```
